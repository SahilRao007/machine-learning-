## random forest tree and depicting the trees included

```python
import pandas as pd 
import numpy as np 
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt 
import seaborn as sns; sns.set_theme(style='darkgrid')
```

```python
iris=load_iris()
```

```python
data=iris.data
target=iris.target
```

```python
iris.target_names
```

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data,target)
```

```python
clf=RandomForestClassifier() #default is 100 sample trees 
```

```python
clf.fit(X_train,y_train)
```

```python
print(clf.base_estimator_)
print(clf.classes_) #output classes 
print(clf.estimator_params)
```

```python
y_pred=clf.predict(X_test)
clf.score(X_test,y_test)
```

```python
from sklearn import tree 
```

```python
plt.figure(figsize=(11,5))
tree.plot_tree(clf.estimators_[1],filled=True); #2nd tree from random tree forest 
```

```python
#ploting first 5 trees
plt.figure(figsize=(10,30))
for i in range(5):
    plt.subplot(5,1,i+1)
    tree.plot_tree(clf.estimators_[i],filled=True)
    plt.title(f"Tree no {i+1}")
   
```

```python
#for simple tree representation 
for i in range(5):
    print(tree.export_text(clf.estimators_[i]))
```
