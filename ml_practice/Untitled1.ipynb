```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set()
```

```python
from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=300, centers=4,cluster_std=0.6,random_state=1)
plt.scatter(X[:,0],X[:,1],s=50)
```

```python
rng=np.random.RandomState(3)
i=rng.permutation(X.shape[0])[:4]
X[i[3]]
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4)
km.fit(X)
y_pred=km.predict(X)
centers=km.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1],s=50)
plt.scatter(centers[:,0],centers[:,1],marker='*')
```

```python
y_pred=km.predict(X)
y_pred
X=np.column_stack((X,y_pred))
```

```python
X=pd.DataFrame(X)
df1=X[X[2]==0]
df2=X[X[2]==1]
df3=X[X[2]==2]
df4=X[X[2]==3]
```

```python
plt.scatter(df1[0],df1[1],color='red')
plt.scatter(df2[0],df2[1],color='purple')
plt.scatter(df3[0],df3[1],color='blue')
plt.scatter(df4[0],df4[1],color='green')
plt.scatter(centers[:,0],centers[:,1],marker='*',color='cyan')
```
