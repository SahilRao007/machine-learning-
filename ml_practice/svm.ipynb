## creating a svc and ploting 

```python
import numpy as np 
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X,y=make_blobs(n_samples=40,centers=2,random_state=20)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='RdYlBu')
```

```python
##creating a support vector classifier
```

```python
clf=SVC(kernel='linear',C=1000)
clf.fit(X,y)
```

```python
clf.support_vectors_ #these are the support vectors that lie closest to the 
#decision boundary and crucial for defing the decision boundary
```

## plotting the svc

```python
plt.scatter(X[:,0],X[:,1],c=y)
ax=plt.gca()
xlim=ax.get_xlim()
ylim=ax.get_ylim()
print(xlim,ylim)
```

```python
#creating grid to evaluate model 
xx=np.linspace(xlim[0],xlim[1],40) #creatign range of 40 space in between 
yy=np.linspace(ylim[0],ylim[1],40)
YY,XX=np.meshgrid(yy,xx)
```

```python
print(XX.shape,YY.shape)


plt.scatter(XX,YY,alpha=0.4,edgecolors='k') #kmeans black
plt.grid(c='black') 
```

### DECISION FUNCTION GUIDE
1. decision_function is a method used in many machine learning libraries, particularly in classifiers, to compute the distance of each sample from the decision boundary. It is often used in binary classification tasks.

2. Positive values indicate one class, negative values indicate the other class, and the magnitude of the value indicates the confidence in the prediction.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

# Load the iris dataset
iris = datasets.load_iris()
Xval = iris.data
yval = iris.target
print(dir(iris))
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(Xval, yval, test_size=0.2, random_state=42)

# Create a Support Vector Classifier
svmm = SVC(kernel='linear')

# Train the classifier
svmm.fit(X_train, y_train)

# Get the decision function values for the test set
decision_values = svmm.decision_function(X_test)

print("Decision function values:\n", decision_values[:5])
```
